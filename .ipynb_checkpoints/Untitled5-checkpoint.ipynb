# Laboratorijas darbs - skaitliskaa integreeshana

Ielaadeeju vajadziigaas darbiibas un savu funkciju darbam

```python
import sys
sys.path.append('/usr/local/anaconda3/lib/python3.6/site-packages')

from numpy import random
from scipy.special import j1
```

Noteicu cik daudz punktus N vajadzeetu lai noteiktu funkcijas laukumu

```python
N = 1000
a = 0
b = 2
```

Noteicu x un y ar funkciju, kas gjeneree ar jebkaadiem skaitljiem

```python
x = random.uniform(a,b,N)
y = random.uniform(a,b,N)
```

Ievietoju vajadziigaas komandas, lai izveidotu grafiku, kaa arii noteicu kur punkti iekraasosies zaljaa un kur sarkanaa krasaa (zaljaa ir tie, kas atrodas grafika laukumaa, tachu sarkanie ir aarpus grafika - taadaa veidaa ar zaljajiem punkiem nosakot funkcijas laukumu)

```python
from matplotlib import pyplot as plt
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Funkcija un taas integraalis (laukums starp funkciju un x ass)')

N1 = 0
for i in range(N):
    if y[i] < j1(2*x[i]):
        plt.plot(x[i],y[i],'go')
        N1 = N1 + 1
    else:
        plt.plot(x[i],y[i],'ro')

S_zinaamais = (b-a) * (b-a)
S_nezinaamais = 1. * S_zinaamais * N1 / N
print(S_nezinaamais)

plt.show()
```
